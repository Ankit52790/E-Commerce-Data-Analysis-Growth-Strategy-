```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load data (Assuming CSV files are structured correctly)
task1_df = pd.read_csv("DT Ecommerce Growth Hacker Assignment - Task1.csv")
task2_df = pd.read_csv("DT Ecommerce Growth Hacker Assignment - Task2.csv")

```

```python
# Check available columns in task2_df
print("Available columns in task2_df:", task2_df.columns)

```

```python
# Attempt to rename a relevant column to 'Sales' if it exists
possible_sales_columns = ["Units Sold", "Total Sales", "Revenue"]
for col in possible_sales_columns:
    if col in task2_df.columns:
        task2_df.rename(columns={col: "Sales"}, inplace=True)
        break

```

```python
# Profitability Breakdown
profit_metrics = {"Gross Margin": 15, "Amazon Fee": 4.5, "Net Margin": 10.5}
plt.figure(figsize=(6, 4))
sns.barplot(x=list(profit_metrics.keys()), y=list(profit_metrics.values()), palette="coolwarm")
plt.title("Profitability Breakdown per Unit ($)")
plt.ylabel("Amount ($)")
plt.show()

```

```python
# Scenario Impact on Profit
scenarios = ["Base Case", "Sales Drop (4000 Units)", "Fixed Cost Increase", "Price Increase ($35)", "Tax Increase (25%)"]
profits = [41000, 30500, 36000, 62250, 30750]
plt.figure(figsize=(8, 5))
sns.lineplot(x=scenarios, y=profits, marker="o", color="b")
plt.xticks(rotation=20)
plt.title("Impact of Different Scenarios on Profit ($)")
plt.ylabel("Total Profit ($)")
plt.show()
```

```python

# Break-even Analysis
cac_values = list(range(5, 16))  # Simulating CAC values from $5 to $15
break_even_profit = [(10.5 - cac) * 5000 for cac in cac_values]  # (Net Margin - CAC) * Units Sold
plt.figure(figsize=(7, 4))
sns.scatterplot(x=cac_values, y=break_even_profit, color="red")
plt.axhline(0, color="black", linestyle="--")
plt.title("Break-even CAC Analysis")
plt.xlabel("Customer Acquisition Cost ($)")
plt.ylabel("Profit ($)")
plt.show()


```

```python
# Sales & Revenue Distribution
if "Sales" in task2_df.columns:
    plt.figure(figsize=(7, 4))
    sns.histplot(task2_df["Sales"], bins=10, kde=True, color="green")
    plt.title("Sales & Revenue Distribution")
    plt.xlabel("Units Sold")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Column 'Sales' not found in task2_df. Check dataset structure.")
```
